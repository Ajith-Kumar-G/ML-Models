```python
#web scraping of a html table for our dataset 
import pandas as pd
```

```python
df = pd.read_html('/content/DailyWiseMarketArrivals.html')[-1]
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.to_csv('market_arrivals.csv')
```

```python
df.info()
```

```python
df.rename(columns = {'Date':'ds','Modal Price (Rs/q)':'y'}, inplace = True)
```

```python
df.info()
```

```python
df.columns
```

```python
df_new = df[['ds','y']] #slicing of data
```

```python
df_new
```

```python
df_new.info()
```

```python
#get the columns in the required format #datetime64 and int64
df_new['ds'] = pd.to_datetime(df_new['ds'])
```

```python
#convert to int64
df_new['y'] = df_new['y'].astype(int)
```

```python
df_new.info()
```

```python
df_new
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(23,10))
plt.scatter(df_new['ds'],df_new['y'], c ='green', cmap= 'Greens')
plt.show()
import seaborn as sns
plt.figure(figsize = (20,13))
sns.scatterplot(x = 'ds', y = 'y', data = df_new, color = 'blue' )
```

```python
import numpy as np 
df_new['y'] = np.where(df_new['y']>2750, np.nan, df_new['y']) #setting outliers 
df_new
```

```python
df_new.isnull().sum()
```

```python
import fbprophet 
op = fbprophet.Prophet(changepoint_prior_scale=1) #makes our model more robust and graph fits nicely  
op.fit(df_new)
```

```python
forecast = op.make_future_dataframe(periods = 365 , freq = 'D') #forecasts for the next 365 days
forecast
```

```python
forecast = op.predict(forecast)
```

```python
plt.figure(figsize=(23,10))
op.plot(forecast,xlabel='Date',ylabel='Prices')
plt.show()
```
